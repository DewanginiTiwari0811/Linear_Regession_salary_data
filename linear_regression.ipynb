```python
#target value
x= data['YearsExperience']#slicing
x=np.array(data[['YearsExperience']])  #array to convert any data in to array we use this and for converting into 2D anD 3D array we use square brackets accordingly like for 2D( [[]]).
y=data['Salary']
y=np.array(data['Salary'])
```

```python
print(type(y))
```

```python
#y is a label because it is in series form ... it is not mandatory to take y variable as one-dimensional
from sklearn.model_selection import train_test_split
xtrain, xtest, ytrain, ytest = train_test_split(x,y,train_size=0.8,random_state =0)
```

```python
#build a model
from sklearn.linear_model import LinearRegression
model = LinearRegression()
#model fitting
model.fit(xtrain,ytrain)
#x test 
#y test
ypred= model.predict(xtest) #for predicting y value(the data we trained)

#mean squared error
from sklearn.metrics import mean_squared_error
mse= mean_squared_error(ytest, ypred)

#r2 square
from sklearn.metrics import r2_score
score = r2_score(ytest, ypred)
score*100
```

```python
#draw the line pf regression(training samples)
plt.scatter(xtrain,ytrain, color='red' , label=" actual data", s=120)
plt.scatter(xtrain,model.predict(xtrain), color='blue', label='Predicted data', s= 120)
plt.plot(xtrain,model.predict(xtrain), color='black', label='line of regression', linestyle= 'dotted')
plt.legend()
plt.show()
```

```python
#draw the line pf regression(testing samples samples)
plt.scatter(xtest,ytest, color='red' , label=" actual data", s=12)
plt.scatter(xtest,ypred, color='blue', label='Predicted data', s= 12)
plt.plot(xtest,ypred, color='black', label='line of regression', linestyle= 'dotted')
plt.legend()
plt.show()
```
